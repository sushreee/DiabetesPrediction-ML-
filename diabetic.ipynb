```python
ip=data.drop(['Outcome'],axis=1)
```

```python
ip
```

```python
op=data['Outcome']
```

```python
data.shape
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ip,op,test_size=0.1)
```

```python
x_train.shape
```

```python
x_train
```

# logistic regression

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
alg1=LogisticRegression()
alg1.fit(x_train,y_train)
```

```python
yp=alg1.predict(x_test)
```

```python
yp
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,yp)
cm
```

```python
cm1=ConfusionMatrixDisplay(cm)
cm1.plot()
plt.show()
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp)
recall=recall_score(y_test,yp)
precision=precision_score(y_test,yp)
f1=f1_score(y_test,yp)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

# naive bayes

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
yp1=nb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc=accuracy_score(y_test,yp1)
recall=recall_score(y_test,yp1)
precision=precision_score(y_test,yp1)
f1=f1_score(y_test,yp1)
```

```python
print(acc)
print(recall)
print(precision)
print(f1)
```

# K nearest neighbour algorithm

```python
#knn algo
from sklearn.neighbors import KNeighborsClassifier
alg2=KNeighborsClassifier(n_neighbors=7)
alg2.fit(x_train,y_train)
```

```python
yp2=alg2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
acc2=accuracy_score(y_test,yp2)
recall2=recall_score(y_test,yp2)
precision2=precision_score(y_test,yp2)
f2=f1_score(y_test,yp2)
```

```python
print(acc2)
print(recall2)
print(precision2)
print(f2)
```

# Support Vector System Algorithm

```python
from sklearn import svm
sv=svm.SVC(kernel='rbf',C=100,gamma=0.01)
sv.fit(x_train,y_train)
```

```python
yp3=sv.predict(x_test)
```
